```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
df = pd.read_csv("titanic task 2/train.csv")
df
```

```python
rows,columns = df.shape
print("number of rows: ",rows)
print("number of coloumns:" ,columns)
```

```python
#data description
df.describe()
```

```python
#age distribution onboard
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'],kde = True,color="maroon")
plt.title('Age Distribution onboard ')

```

```python
#Gender distribution onboard
plt.figure(figsize=(6, 4))
df['Sex'].value_counts().plot(kind='bar', color=['cadetblue', 'salmon'], edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number')
plt.show()

```

```python
# Age Distribution by Gender and Passenger Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df, palette='Paired')
plt.title('Age Distribution by Gender and Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Fare Distribution by Passenger Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='Set2')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Create Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+'])

# Survival Analysis by Age Group
plt.figure(figsize=(8, 6))
sns.countplot(x='AgeGroup', hue='Survived', data=df, palette="Set1" , edgecolor='black')
plt.title('Survival Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Survival Analysis by Gender and Passenger Class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Sex', data=df, palette='Paired', edgecolor='black')
plt.title('Survival Count by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```
